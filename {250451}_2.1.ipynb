```python
import  pandas as pd
import numpy as np

from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('Dataset_1 - Sheet1.csv')
```

```python
data
```

```python
house_area = data['House Area']
no_of_rooms = data['Number of rooms']
price = data['Price']
```

```python
x0 = np.random.rand(1)
x1 = np.random.rand(1)
x2 = np.random.rand(1)
```

```python
for i in range(100):
  pred_price = x0 + x1*house_area + x2*no_of_rooms
  loss_matrix = price - (x0 + x1*house_area + x2*no_of_rooms)
  loss = np.sum(np.square(loss_matrix))
  x0_grad = -2*np.sum(loss_matrix)
  x1_grad = -2*np.dot(house_area, loss_matrix)
  x2_grad = -2*np.dot(no_of_rooms, loss_matrix)
  lr = 1e-10
  x0 -= lr*x0_grad
  x1 -= lr*x1_grad
  x2 -= lr*x2_grad
  print(i, loss)
```

```python
print(x0, x1, x2)
```

```python
comp = pd.concat([pred_price, price], axis=1)
```
