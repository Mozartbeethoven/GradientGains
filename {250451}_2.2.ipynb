```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('Dataset_2 - Sheet1.csv')
```

```python
data_replaced = data.copy()


gender_mapping = {'M': 1, 'F': 0}
department_mapping = {'Finance':1, 'Software' : 2, 'HR' : 0 }


data_replaced['Gender'] = data_replaced['Gender'].replace(gender_mapping)
data_replaced['Department'] = data_replaced['Department'].replace(department_mapping)


data_replaced
```

```python
y = data['Team']
num_A = np.sum(y == 'A')
num_B = np.sum(y == 'B')

initial_gini = 1 - (num_A / len(y))**2 - (num_B / len(y))**2
initial_gini
```

```python
def gini_impurity(feature, threshold):

  left_indices = feature.index[feature <= threshold]
  right_indices = feature.index[feature > threshold]


  y_left = y.loc[left_indices]
  y_right = y.loc[right_indices]

  total_len = len(feature)


  num_A_left = np.sum(y_left == 'A')
  num_B_left = np.sum(y_left == 'B')
  len_left = len(y_left)

  gini_left = 0
  if len_left > 0:
    prob_A_left = num_A_left / len_left
    prob_B_left = num_B_left / len_left
    gini_left = 1 - prob_A_left**2 - prob_B_left**2


  num_A_right = np.sum(y_right == 'A')
  num_B_right = np.sum(y_right == 'B')
  len_right = len(y_right)

  gini_right = 0
  if len_right > 0:
    prob_A_right = num_A_right / len_right
    prob_B_right = num_B_right / len_right
    gini_right = 1 - prob_A_right**2 - prob_B_right**2


  gini_split = (len_left / total_len) * gini_left + (len_right / total_len) * gini_right
  return gini_split
```

```python
def info_gain(feature, threshold, parent_gini):
  gini_impurity_after = gini_impurity(feature, threshold)
  return parent_gini - gini_impurity_after
```

```python
Income = data_replaced['Income']
for i in range(len(Income)):
  Income[i] = int(Income[i])
```

```python
best_income_split = -10
for i in range(len(Income)):
  best_income_split = max(best_income_split, info_gain(Income, Income[i], initial_gini))

print(best_income_split)
```

```python
Gender = data_replaced['Gender']
best_gender_split = -10
for i in range(len(Gender)):
  best_gender_split = max(best_gender_split, info_gain(Gender, Gender[i], initial_gini))

print(best_gender_split)
```

```python
Department = data_replaced['Department']
best_department_split = -10
for i in range(len(Department)):
  best_department_split = max(best_department_split, info_gain(Department, Department[i], initial_gini))

print(best_department_split)
```

```python
first_split = max(best_income_split, best_gender_split, best_department_split)
print(first_split)
```

```python
data_second = data_replaced.loc[Gender==1]
data_second
```

```python

y_second_parent = y.loc[data_second.index]
num_A_second_parent = np.sum(y_second_parent == 'A')
num_B_second_parent = np.sum(y_second_parent == 'B')
len_second_parent = len(y_second_parent)

gini_second_layer_parent = 0
if len_second_parent > 0:
  prob_A_second_parent = num_A_second_parent / len_second_parent
  prob_B_second_parent = num_B_second_parent / len_second_parent
  gini_second_layer_parent = 1 - prob_A_second_parent**2 - prob_B_second_parent**2

print(f"Gini impurity for the second layer parent (Gender=1): {gini_second_layer_parent:.10f}")
```

```python
Income = data_second['Income']
best_income_split_second = -10
for threshold_val in Income.unique():
  current_info_gain = info_gain(Income, threshold_val, gini_second_layer_parent)
  best_income_split_second = max(best_income_split_second, current_info_gain)
  print(f"Info gain for threshold {threshold_val}: {current_info_gain:.10f} (Current best: {best_income_split_second:.10f})")

print(f"Overall best income split for second layer: {best_income_split_second:.10f}")
```

```python
Department = data_second['Department']
best_department_split_second = -10
for threshold_val in Department.unique():
  current_info_gain = info_gain(Department, threshold_val, gini_second_layer_parent)
  best_department_split_second = max(best_department_split_second, current_info_gain)
  print(f"Info gain for threshold {threshold_val}: {current_info_gain:.10f} (Current best: {best_department_split_second:.10f})")
print(f"Overall best department split for second layer: {best_department_split_second:.10f}")
```

```python

y_third_parent = y.loc[data_third.index]
num_A_third_parent = np.sum(y_third_parent == 'A')
num_B_third_parent = np.sum(y_third_parent == 'B')
len_third_parent = len(y_third_parent)

gini_third_layer_parent = 0
if len_third_parent > 0:
  prob_A_third_parent = num_A_third_parent / len_third_parent
  prob_B_third_parent = num_B_third_parent / len_third_parent
  gini_third_layer_parent = 1 - prob_A_third_parent**2 - prob_B_third_parent**2

print(f"Gini impurity for the third layer parent (Income>=39000 from Gender=1): {gini_third_layer_parent:.10f}")
```

```python
Department = data_third['Department']
best_department_split_third = -10
for threshold_val in Department.unique():
  current_info_gain = info_gain(Department, threshold_val, gini_third_layer_parent)

  best_department_split_third = max(best_department_split_third, current_info_gain)
  print(f"Info gain for threshold {threshold_val}: {current_info_gain:.10f} (Current best: {best_department_split_third:.10f})")
print(f"Overall best department split for third layer: {best_department_split_third:.10f}")
```
